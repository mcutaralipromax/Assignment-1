```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

Bar Chart

```python
plt.figure(figsize=(17,10))

plt.subplot(2,2,1)
sns.barplot(x=data['year'].value_counts().index, y=data['major_category'].value_counts())
plt.ylabel('Kejahatan Berat')
plt.xlabel('Tahun')
plt.xticks(rotation=90)
plt.title('Kejahatan berat dari Tahun 2008 sampai 2016')

plt.show()
```

Jika melihat tabel tersebut maka bisa disimpulkan kejahatan sering terjadi pada tahun 2016 dan jarang terjadi ada tahun 2009

Saya akan membandingkan kejahatan di daerah Ealing dengan Sutton di tahun 2016. Karena pada tahun tersebut kejahatan banyak terjadi

Sebelum menvisualisasikannya lebih baik melakukan filtering pada data agar mudah untuk mengolahnya

```python
data.head()
```

```python
dataEaling = data[(data.borough == "Ealing") & (data.year == 2016)]
dataEaling
```

```python
dataSutton = data[(data.borough == "Sutton") & (data.year == 2016)]
dataSutton
```

```python
fig = plt.figure()
ax = fig.add_axes ([0,0,1,1])
x = (dataEaling.major_category)
y = (dataEaling.value)
ax.bar(x,y)
plt.xlabel('Jenis Kejahatan')
plt.ylabel('value')
plt.title('Kejahatan di Ealing tahun 2016')
plt.xticks(rotation = 'vertical')

fig = plt.figure()
ax = fig.add_axes ([0,0,1,1])
x = (dataSutton.major_category)
y = (dataSutton.value)
ax.bar(x,y)
plt.xlabel('Jenis Kejahatan')
plt.ylabel('value')
plt.title('Kejahatan di Sutton tahun 2016')
plt.xticks(rotation = 'vertical')
plt.show()
```
